## Importing the loan payments data:

```python
import pandas as pd
from loans_data_eda import DataTransform
from loans_data_eda import DataFrameInfo
from loans_data_eda import DataFrameTransform
from loans_data_eda import Plotter
```

```python
loan_payments_df = pd.loan_payments_df = pd.read_csv('loan_payments_data.csv')
```

## Converting columns to the correct format:

```python
# Initialise DataTransform class with imported dataframe 
data_type_transformer = DataTransform(loan_payments_df)

# Transform datatypes of columns where necessary
data_type_transformer.transform_all()
format_transformed_df = data_type_transformer.transform_all()

print(format_transformed_df.info())
```

## Extracting information from the DataFrame and it's columns:

```python
# Initialise DataFrameInfo class with datatype corrected dataframe 
df_info = DataFrameInfo(format_transformed_df) 

# Extract information about columns and dataframe 
df_info.describe_columns_and_datatypes()
print(df_info.extract_statistical_values())
df_info.count_category_distinct_values()
df_info.print_dataframe_shape()
df_info.percentage_of_nulls()
```

## Removing/imputing missing values in the data:

```python
# Initialise DataFrameTransform class with datatype corrected dataframe 
dataframe_transformer = DataFrameTransform(format_transformed_df) 

# Check null percentages to identify columns with missing values 
dataframe_transformer.check_percentage_of_nulls() 
```

The columns with a considerable percentage of missing values (mnths_since_last_delinq, mths_since_last_record, next_payment_date, mths_since_last_major_derog) should be removed as there is too much data missing to be able to impute values. 



```python
# Apply transformations to columns with missing values 
dataframe_transformer.transform_all_nulls() 
null_transformed_df = dataframe_transformer.transform_all_nulls() 

# Check there are no missing values 
dataframe_transformer.check_percentage_of_nulls() 
```

```python
# Initialise Plotter class 
dataframe_plotter = Plotter() 

# Plot barchart to visualise the removal/imputation of missing values 
dataframe_plotter.visualise_nulls_removal(format_transformed_df, null_transformed_df) 
```

## Performing transformations on skewed columns:

### Identifying skewed columns:

```python
# Select numerical columns from dataframe and exclude columns for which the values must not be changed
numeric_columns = null_transformed_df.select_dtypes(include=[int, float]).columns.tolist()
columns_to_exclude = [
    'id',
    'member_id',
    'policy_code'
]
columns_to_investigate_skewness = [col for col in numeric_columns if col not in columns_to_exclude]

# Calculate skewness of numerical columns 
skewness = null_transformed_df[columns_to_investigate_skewness].skew()

print(f"Skewness of numerical columns: \n{skewness}")
```

```python
# Filter columns based on skewness threshold 
skewed_columns = skewness[skewness > 0.5].index

# Plot barcharts to compare effect of log and yeo-johnson transformations on distribution of data
dataframe_plotter.plot_skew_transformations(null_transformed_df, skewed_columns)
```

The above histogram and the skewness values following log transformation show that the previously positively skewed columns have become negatively skewed as a result of the transformation, while Yeo-Johnson transformation results in normal distribution for a large number of columns.

The following columns are still skewed above the threshold of 0.5 and so will not be transformed: 
* delinq_2yrs
* out_prncp
* out_prncp_inv
* total_rec_late_fee
* recoveries
* collection_recovery_fee
* collections_12_mths_ex_med


### Transforming skewed columns with Yeo_Johnson transform:

```python
# Exclude columns that remained skewed following transformations
columns_not_to_transform = [
    'delinq_2yrs',
    'out_prncp',
    'out_prncp_inv',
    'total_rec_late_fee',
    'recoveries',
    'collection_recovery_fee',
    'collections_12_mths_ex_med'
    ]
columns_to_transform = [col for col in skewed_columns if col not in columns_not_to_transform]

# Apply yeo-johnson transformation on skewed columns 
skew_transformed_df = dataframe_transformer.transform_skewed_columns(null_transformed_df, columns_to_transform)
```

### Check Yeo-Johnson transformation has been applied to data:

```python
# Calculate skewness of transformed columns to check columns have been transformed
skewness = skew_transformed_df[skewed_columns].skew()

print(f"Skewness of numerical columns: \n{skewness}")
```

## Removing outliers from the data:

Identify outliers in columns by visualisation using boxplots:

```python
# Plot boxplots of normally distributed columns 
dataframe_plotter.visualise_outliers(skew_transformed_df, columns_to_transform)
```

The correlation matrix shows high collinearity between:

### Remove overly correlated columns:

Remove columns with correlation above 0.9. 

loan_amount and funded_amount have high collinearity and due to the imputation for 5% of missing values in funded_amount, I shall keep loan_amount and drop funded_amount

```python
# Create list of columns to drop from data due to collinearity
collinear_columns_to_drop = [
    'id',
    'funded_amount',
    'funded_amount_inv',
    'out_prncp_inv',
    'total_payment_inv',
    'total_rec_prncp'
]

# Drop highly collinear columns 
corr_transformed_df = dataframe_transformer.drop_collinear_columns(outlier_transformed_df, collinear_columns_to_drop)
```
