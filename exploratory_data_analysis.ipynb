# Milestone 3: Exploratory Data Analysis 

```python
import pandas as pd
from loans_data_eda import DataTransform
from loans_data_eda import DataFrameInfo
from loans_data_eda import DataFrameTransform
from loans_data_eda import Plotter
```

## Importing the loan payments data:

```python
loan_payments_df = pd.loan_payments_df = pd.read_csv('loan_payments_data.csv')
```

## Converting columns to the correct format:

```python
data_type_transformer = DataTransform(loan_payments_df) # Initialise DataTransform class with imported dataframe 
format_transformed_df = data_type_transformer.transform_all() # Transform data types of columns where necessary and store result in format_transformed_df

print(format_transformed_df.info())
```

## Extracting information from the DataFrame and it's columns:

```python
df_info = DataFrameInfo(format_transformed_df) # Initialise DataFrameInfo class with data type corrected dataframe 

df_info.describe_columns_and_datatypes() # Extracts data type and non-null count of each column in dataframe 
print(df_info.extract_statistical_values()) # Calculates count, mean, std, min, max, etc for each column
df_info.count_category_distinct_values() # Calculates the number of each unique value for category type columns 
df_info.print_dataframe_shape() # Calculates number of rows and columns in dataframe 
```

## Removing/imputing missing values in the data:

```python
dataframe_transformer = DataFrameTransform(format_transformed_df) # Initialise DataFrameTransform class with data type corrected dataframe 
dataframe_transformer.check_percentage_of_nulls(format_transformed_df) # Check null percentages to identify columns with missing values 
```

The columns with a considerable percentage of missing values (mnths_since_last_delinq, mths_since_last_record, next_payment_date, mths_since_last_major_derog) should be removed as there is too much data missing to be able to impute values. 



```python
null_transformed_df = dataframe_transformer.transform_all_nulls() # Apply transformations to columns with missing values 
dataframe_transformer.check_percentage_of_nulls(null_transformed_df) # Check there are no missing values 
```

```python
dataframe_plotter = Plotter() # Initialise Plotter class 
dataframe_plotter.visualise_nulls_removal(format_transformed_df, null_transformed_df) # Plot barchart to visualise the removal/imputation of missing values 
```

## Performing transformations on skewed columns:

### Identifying skewed columns:

```python
numeric_columns = null_transformed_df.select_dtypes(include=[int, float]).columns.tolist() # Select numerical columns from dataframe
columns_to_exclude = [
    'id',
    'member_id',
    'policy_code'
] # Create list of columns to exclude from skew transformations
columns_to_investigate_skewness = [col for col in numeric_columns if col not in columns_to_exclude]

skewness = null_transformed_df[columns_to_investigate_skewness].skew() # Calculate skewness of numerical columns 

print(f"Skewness of numerical columns: \n{skewness}")
```

```python
skewed_columns = skewness[skewness > 0.5].index # Filter columns based on skewness threshold 
dataframe_plotter.plot_skew_transformations(null_transformed_df, skewed_columns) # Plot barcharts to compare effect of log and yeo-johnson transformations on distribution of data
```

The above histogram and the skewness values following log transformation show that the previously positively skewed columns have become negatively skewed as a result of the transformation, while Yeo-Johnson transformation results in normal distribution for a large number of columns.

The following columns are still skewed above the threshold of 0.5 and so will not be transformed: 
* delinq_2yrs
* out_prncp
* out_prncp_inv
* total_rec_late_fee
* recoveries
* collection_recovery_fee
* collections_12_mths_ex_med


### Transforming skewed columns with Yeo_Johnson transform:

```python
columns_not_to_transform = [
    'delinq_2yrs',
    'out_prncp',
    'out_prncp_inv',
    'total_rec_late_fee',
    'recoveries',
    'collection_recovery_fee',
    'collections_12_mths_ex_med'
    ] # Create list of columns that remained skewed following transformations
columns_to_transform = [col for col in skewed_columns if col not in columns_not_to_transform] # Filter out columns to be excluded 
skew_transformed_df = dataframe_transformer.transform_skewed_columns(null_transformed_df, columns_to_transform) # Apply yeo-johnson transformation on skewed columns 
```

### Check Yeo-Johnson transformation has been applied to data:

```python
skewness = skew_transformed_df[skewed_columns].skew() # Calculate skewness of transformed columns to check columns have been transformed

print(f"Skewness of numerical columns: \n{skewness}")
```

## Removing outliers from the data:

Identify outliers in columns by visualisation using boxplots:

```python
dataframe_plotter.visualise_outliers(skew_transformed_df, columns_to_transform) # Plot boxplots of normally distributed columns 
```

Removing outliers:

```python
outlier_transformed_df = dataframe_transformer.remove_outliers(skew_transformed_df, columns_to_transform) # Remove outliers from columns 
dataframe_plotter.visualise_outliers(outlier_transformed_df, columns_to_transform) # Check outliers have been removed with boxplots for transformed dataframe
```

## Dropping overly correlated columns:

### Identify overly correlated columns by plotting a correlation matrix to visualise collinearity:

```python
dataframe_plotter.plot_correlation_matrix(outlier_transformed_df[numeric_columns]) # Plot correlation matrix of numerical columns for dataframe with outliers removed
```

The correlation matrix shows high collinearity between:
* id and member_id
* loan_amount, funded_amount, funded_amount_inv and instalment
* out_prncp and out_prncp_inv
* total_payment, total_payment_inv and total_rec_prncp

### Remove overly correlated columns:

Remove columns with correlation above 0.9. 

loan_amount and funded_amount have high collinearity and due to the imputation for 5% of missing values in funded_amount, I shall keep loan_amount and drop funded_amount

```python
collinear_columns_to_drop = [
    'id',
    'funded_amount',
    'funded_amount_inv',
    'out_prncp_inv',
    'total_payment_inv',
    'total_rec_prncp'
] # Create list of columns to drop from data due to collinearity
corr_transformed_df = dataframe_transformer.drop_collinear_columns(outlier_transformed_df, collinear_columns_to_drop) # Drop highly collinear columns 
```
